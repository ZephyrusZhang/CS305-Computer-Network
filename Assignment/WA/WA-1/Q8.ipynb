```python
import numpy as np
from matplotlib import pyplot as plt


u = [100 * 1000, 600 * 1000, 4 * 1000000]
us = 30000000
F = 15 * (2 ** 30)
d = 2000000
N = np.linspace(1, 1000, num=114514)
```

```python
D = np.maximum(N * F / us, F / d)
plt.plot(N, D)
plt.show()
```

```python
title = ['$u=100Kps$', '$u=600Kps$', '$u=4Mbps$']

for i in range(len(u)):
    D = np.maximum(np.maximum(F / us, F / d), N * F / (us + N * u[i]))
    plt.plot(N, D)
    plt.title(title[i])
    plt.show()
```
